INSTALL AND IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

LOAD DATASET

```python
df = pd.read_csv("Housing.csv")   # Use correct path if needed
df.head()
```

```python
df.columns
```

TASK 1: Price Range Analysis

```python
df["price_lakhs"] = df["price"] / 100000
```

```python
bins = [0, 25, 50, 75, 100, np.inf]
labels = ["0-25", "26-50", "51-75", "76-100", ">100"]

df["price_category"] = pd.cut(df["price_lakhs"], bins=bins, labels=labels)
```

```python
price_counts = df["price_category"].value_counts().sort_index()
print(price_counts)
```

Line Chart

```python
plt.figure()
price_counts.plot(kind="line", marker="o")
plt.title("Number of Houses by Price Range")
plt.xlabel("Price Range (Lakhs)")
plt.ylabel("Number of Houses")
plt.show()
```

TASK 2: Average Price AC vs Non-AC

```python
df["airconditioning"].unique()
```

```python
ac_avg = df.groupby("airconditioning")["price_lakhs"].mean()
print(ac_avg)
```

Bar Chart

```python
plt.figure()
ac_avg.plot(kind="bar")
plt.title("Average House Price: AC vs Non-AC")
plt.xlabel("Air Conditioning")
plt.ylabel("Average Price (Lakhs)")
plt.show()
```

TASK 3: Relationship Between Parking & Price

```python
df["parking"].unique()
```

```python
parking_price = df.groupby("parking")["price_lakhs"].mean()
print(parking_price)
```

Plot Relationship

```python
plt.figure()
plt.scatter(df["parking"], df["price_lakhs"])
plt.title("Parking vs House Price")
plt.xlabel("Number of Parking Spaces")
plt.ylabel("Price (Lakhs)")
plt.show()
```

TASK 4: Price Gap AnalysiS

Compare:

<5000 sqft & no prefarea

>5000 sqft & prefarea = yes

```python
group1 = df[(df["area"] < 5000) & (df["prefarea"] == "no")]
group2 = df[(df["area"] > 5000) & (df["prefarea"] == "yes")]
```

```python
avg1 = group1["price_lakhs"].mean()
avg2 = group2["price_lakhs"].mean()

print("Avg Price (<5000 sqft & no prefarea):", avg1)
print("Avg Price (>5000 sqft & prefarea):", avg2)

price_gap = avg2 - avg1
print("Price Gap:", price_gap)
```

BAR CHART (Visualization)

```python
plt.figure()

groups = ["<5000 & No Prefarea", ">5000 & Prefarea"]
values = [avg1, avg2]

plt.bar(groups, values)
plt.title("Price Comparison Based on Area & Preferred Area")
plt.ylabel("Average Price (Lakhs)")
plt.xticks(rotation=15)

plt.show()
```

Houses with area greater than 5000 sqft located in preferred areas have significantly higher average prices compared to smaller houses in non-preferred areas. The calculated price gap indicates the strong influence of both area and location on housing price
